```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
L = float(input("Enter the length of the beam: "))
load_type = input("Enter the type of load (concentrated or uniformly distributed): ")

if load_type.lower() == 'concentrated':
    P = float(input("Enter the point load: "))
    a = float(input("Enter the distance from left end to the point load: "))
    # x coordinates
    x = np.linspace(0, L, 500)
    # Shear force
    V = np.piecewise(x, [x < a, x >= a], [lambda x: P, lambda x: 0])
    # Bending moment
    M = np.piecewise(x, [x < a, x >= a], [lambda x: P * x, lambda x: P * a])

elif load_type.lower() == 'uniformly distributed':
    w = float(input("Enter the value of the uniformly distributed load: "))
    # x coordinates
    x = np.linspace(0, L, 500)
    # Shear force
    V = w * (L - x)
    # Bending moment
    M = 0.5 * w * (L - x)**2

else:
    print("Invalid load type. Please enter either 'concentrated' or 'uniformly distributed'.")
```

```python
plt.figure(figsize=(12, 6))
```

```python
plt.subplot(2, 1, 1)
plt.plot(x, V, label='Shear Force')
plt.xlabel('x')
plt.ylabel('V (kN)')
plt.title('Shear Force Diagram')
plt.grid(True)
```

```python
plt.subplot(2, 1, 2)
plt.plot(x, M, label='Bending Moment')
plt.xlabel('x')
plt.ylabel('M (kNm)')
plt.title('Bending Moment Diagram')
plt.grid(True)
```

```python
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# User inputs
L = float(input("Enter the length of the beam: "))
load_type = int(input("Enter the type of load (1 for concentrated, 2 for uniformly distributed, 3 for mixed): "))

if load_type == 1:
    n = int(input("Enter the number of concentrated loads: "))
    P = []
    a = []
    for i in range(n):
        P.append(float(input(f"Enter the magnitude of concentrated load {i+1}: ")))
        a.append(float(input(f"Enter the distance from left end to the concentrated load {i+1}: ")))
    # x coordinates
    x = np.linspace(0, L, 500)
    # Shear force
    V = np.piecewise(x, [x < a[i] for i in range(n)], [lambda x, i=i: P[i] for i in range(n)])
    # Bending moment
    M = np.piecewise(x, [x < a[i] for i in range(n)], [lambda x, i=i: P[i] * x for i in range(n)])

elif load_type == 2:
    n = int(input("Enter the number of uniformly distributed loads: "))
    w = []
    a = []
    b = []
    for i in range(n):
        w.append(float(input(f"Enter the magnitude of uniformly distributed load {i+1}: ")))
        a.append(float(input(f"Enter the starting distance from left end to the uniformly distributed load {i+1}: ")))
        b.append(float(input(f"Enter the ending distance from left end to the uniformly distributed load {i+1}: ")))
    # x coordinates
    x = np.linspace(0, L, 500)
    # Shear force
    V = np.piecewise(x, [x < a[i] for i in range(n)], [lambda x, i=i: w[i] * (b[i] - x) for i in range(n)])
    # Bending moment
    M = np.piecewise(x, [x < a[i] for i in range(n)], [lambda x, i=i: 0.5 * w[i] * (b[i] - x)**2 for i in range(n)])

elif load_type == 3:
    n_P = int(input("Enter the number of concentrated loads: "))
    P = []
    a_P = []
    for i in range(n_P):
        P.append(float(input(f"Enter the magnitude of concentrated load {i+1}: ")))
        a_P.append(float(input(f"Enter the distance from left end to the concentrated load {i+1}: ")))
    n_w = int(input("Enter the number of uniformly distributed loads: "))
    w = []
    a_w = []
    b_w = []
    for i in range(n_w):
        w.append(float(input(f"Enter the magnitude of uniformly distributed load {i+1}: ")))
        a_w.append(float(input(f"Enter the starting distance from left end to the uniformly distributed load {i+1}: ")))
        b_w.append(float(input(f"Enter the ending distance from left end to the uniformly distributed load {i+1}: ")))
    # x coordinates
    x = np.linspace(0, L, 500)
    # Shear force
    V = np.piecewise(x, [x < a_P[i] for i in range(n_P)], [lambda x, i=i: P[i] for i in range(n_P)]) + \
        np.piecewise(x, [x < a_w[i] for i in range(n_w)], [lambda x, i=i: w[i] * (b_w[i] - x) for i in range(n_w)])
    # Bending moment
    M = np.piecewise(x, [x < a_P[i] for i in range(n_P)], [lambda x, i=i: P[i] * x for i in range(n_P)]) + \
        np.piecewise(x, [x < a_w[i] for i in range(n_w)], [lambda x, i=i: 0.5 * w[i] * (b_w[i] - x)**2 for i in range(n_w)])

else:
    print("Invalid load type. Please enter 1 for concentrated, 2 for uniformly distributed, or 3 for mixed.")

# Plotting
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(x, V, label='Shear Force')
plt.xlabel('x')
plt.ylabel('V (kN)')
plt.title('Shear Force Diagram')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(x, M, label='Bending Moment')
plt.xlabel('x')
plt.ylabel('M (kNm)')
plt.title('Bending Moment Diagram')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# User inputs
L = float(input("Enter the length of the beam: "))
load_type = int(input("Enter the type of load (1 for concentrated, 2 for uniformly distributed, 3 for mixed): "))

# Initialize arrays for shear force and bending moment
x = np.linspace(0, L, 500)
V = np.zeros_like(x)
M = np.zeros_like(x)

if load_type in [1, 3]:
    n_P = int(input("Enter the number of concentrated loads: "))
    for i in range(n_P):
        P = float(input(f"Enter the magnitude of concentrated load {i+1}: "))
        a_P = float(input(f"Enter the distance from left end to the concentrated load {i+1}: "))
        # Update shear force and bending moment
        V += np.piecewise(x, [x < a_P, x >= a_P], [lambda x: P, lambda x: 0])
        M += np.piecewise(x, [x < a_P, x >= a_P], [lambda x: P * x, lambda x: P * a_P])

if load_type in [2, 3]:
    n_w = int(input("Enter the number of uniformly distributed loads: "))
    for i in range(n_w):
        w = float(input(f"Enter the magnitude of uniformly distributed load {i+1}: "))
        a_w = float(input(f"Enter the starting distance from left end to the uniformly distributed load {i+1}: "))
        b_w = float(input(f"Enter the ending distance from left end to the uniformly distributed load {i+1}: "))
        # Update shear force and bending moment
        V += np.piecewise(x, [x < a_w, x < b_w, x >= b_w], [lambda x: 0, lambda x: w * (x - a_w), lambda x: w * (b_w - a_w)])
        M += np.piecewise(x, [x < a_w, x < b_w, x >= b_w], [lambda x: 0, lambda x: 0.5 * w * (x - a_w)**2, lambda x: 0.5 * w * (b_w - a_w)**2 + w * (b_w - a_w) * (x - b_w)])

# Plotting
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(x, V, label='Shear Force')
plt.xlabel('x')
plt.ylabel('V (kN)')
plt.title('Shear Force Diagram')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(x, M, label='Bending Moment')
plt.xlabel('x')
plt.ylabel('M (kNm)')
plt.title('Bending Moment Diagram')
plt.grid(True)

plt.tight_layout()
plt.show()
```
